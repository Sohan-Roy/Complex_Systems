# Handling Matrices in Python

## Define a 3 x 3 matrix and display its rows and elements

```python
import numpy as np #python library for mathematical and linear algebra modules

#Let us define a matrix A with rows (1,1,0), (2,7,2), (3,6,1)

A = np.array([[1,1,0],[2,7,2],[3,6,1]])

#syntax created : A[i,j], where i = row index, j = column index
#i,j = 0,1,2,... ; hence nth element has index n-1

print(A[0])
print(A[2])
print(" ")
print("The element in the 3rd row and 2nd column is:", A[2,1])
```

#### Calculation of the determinant and trace of a square matrix :

```python
import numpy as np
A = np.array([[1,1,0],[2,7,2],[3,6,1]])

print("Given matrix =\n ", A)

t = np.trace(A) #calculates trace
print("The trace of the given matrix : ", t)

d = np.linalg.det(A) #calculates determinant
print("The determinant of the given matrix : ",d)
```

#### Products of two or more matrices

```python
import numpy as np
A = np.array([[1,1,0],[2,7,2],[3,6,1]])
B = np.array([[2,3,1],[2,4,2],[1,6,1]]) 
print("Matrix A :\n", A)
print("MAtrix B :\n", B)

C = np.dot(A,B)
print("Product of the matrices A and B:\n", C)
```

#### Finding the eigenvalues and eigenvectors of a matrix : 

```python
import numpy as np
A = np.array([[1,2],[3,-4]])

u,v = np.linalg.eig(A) # u = stores eigenvalues, v = stores eigenvectors

print("Eigenvalues of A are :", u)
```

```python
print(v) #the columns of v represent the eigenvectors
```

```python
w = np.transpose(v)
print("Eigenvectors of A are :")
print(w[0],w[1])
```

### Question : Write a program to create a matrix A as shown below:

#### 5     2     0

### Print the following :
#### (i) The matrix A
#### (ii) The rows of A
#### (iii) The eigenvalues of matrix A
#### (iv) The determinant of matrix A
#### (v) The eigenvectors of matrix A

```python
import numpy as np

a = np.array([[5,2,0],[2,5,0],[-3,4,6]])
print("Matrix A:\n", a)

for i in range(len(a)):
    r = a[i]
    print("The ",i+1,"th row is : ",r)
    
eig, vec = np.linalg.eig(a)
print(" ")
print("The eigenvalurs of matrix (a) are : ", eig)

det = np.linalg.det(a)
print(" ")
print("The determinant of the matrix (a) is : ", det)

vec1=np.transpose(vec)
print("The eigenvectors of (a) are : ", vec1[0], vec1[1], vec1[2])
```

## Creating a graph by calling files that store info about nodes and edges :

nodes.txt contains the nodes of the graph
edges.txt contains the nodes of the graph

```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt

N = np.loadtxt(r"D:\Complex Systems Coding\python file handling\nodes.txt", dtype = int)
G = nx.DiGraph()

G.add_nodes_from(N)

#to access an edge list, we use the following commands
h = nx.read_edgelist(r'D:\Complex Systems Coding\python file handling\edges.txt', create_using=nx.DiGraph(), nodetype=int)

G.add_edges_from(h.edges)
nx.draw_planar(G, node_size = 500, with_labels = True)

#generate the adjacency matrix
adj = nx.to_numpy_array(G) # A_ij =1 for node i ----> j in python
print(adj)
```

### Labelling the nodes

```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt

adj = np.loadtxt(r"D:\Complex Systems Coding\python file handling\Adjacency_mat.txt", dtype='i')

# to create graph, use the following command
G = nx.from_numpy_array(adj, create_using = nx.DiGraph())

N = list(G.nodes())
print(len(N))

# As there are 9 nodes, we can create a node dictionary with 9 labels
label = ["a","b","c","d","e","f","g","h","i"]
labeldict = {}
for i in range(len(N)):
    labeldict[i] = label[i]
nx.draw_circular(G, labels = labeldict, with_labels=True)



```
