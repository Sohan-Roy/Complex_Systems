## 1. Basic python and Matplotlib :

```python
#We want to plot two datasets x and y

import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y = [1,4,9,16,25]
plt.plot(x,y) #by default plot() function uses lines
plt.show()

#plt.savefig("Sampleplot1.png")
```

### Exercise : Find out and execute instructions on how to label the axes and modify the plot "attributes" (marker type, marker size, marker colour, etc.)

```python
#Modify plot attributes 1

import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y = [1,4,9,16,25]
plt.plot(x,y,".",color = 'r') #modified to use dots as markers, and colour it red(r)
plt.xlabel("x-data") #adds label to the x axis
plt.ylabel("y-data") #adds label to the y axis
plt.title("A plot of y vs. x") #adds a title to the plot

plt.show()
```

```python
#Modify plot attributes 2 (adding a legend and changing marker style)

import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y = [1,4,9,16,25]
z = [1,8,27,64,125]
plt.plot(x,y,'ro', label = "y-data") #ro = red circles
plt.plot(x,z,'gs', label = "z-data") #gs = green squares
plt.legend(loc='best') #puts a legend

plt.show()
```

## 2. The Networkx Library

```python
# Plotting an Undirected graph

import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()

nodes = [1,2,3,4,5]
edges = [(1,2),(1,3),(2,4),(2,5),(3,5)]

G.add_nodes_from(nodes)
G.add_edges_from(edges)
nx.draw(G,with_labels=True)

plt.show()
```

### Exercise : Find out and execute instructions on how to modify the graph "attributes" (node type,node size, node colour, etc.)

```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()
G.graph["Name"] = "Sample Graph"
nodes = [1,2,3,4,5]
edges = [(1,2),(1,3),(2,4),(2,5),(3,5)]

G.add_nodes_from(nodes)
G.add_edges_from(edges)
nx.draw_planar(G,with_labels=True,node_color="red",node_size=3000)

plt.show()
```

```python
#Adding Node attributes and Directed Graph

import networkx as nx 
import matplotlib.pyplot as plt 

G=nx.DiGraph() 

#nodes2=['A','B','C','D'] 
#edges2=[('A','B'), ('B','C'), ('C','A'),('A','D')] 
#G.add_nodes_from(nodes2)
#G.add_edges_from(edges2) 

#The above commented lines can also be programmed as follows
G.add_nodes_from("ABCD")
G.add_edges_from(["AB","BC","CA","AD"])

nx.draw(G,with_labels= True) 
plt.show()
```

```python
#Adding Attributes
import networkx as nx 
import matplotlib.pyplot as plt 

G=nx.DiGraph() 

G.add_node("A", age=21, gender="Male")
G.add_node("B", age=22, gender="Female")
G.add_node("C", age=19, gender="Female")
G.add_node("D", age=21, gender="Female")

G.add_edge("A","B", weight=1)
G.add_edge("B","C", weight=0.5)
G.add_edge("C","A", weight=0.5)
G.add_edge("A","D", weight=1)


nx.draw(G,with_labels= True)
plt.show()
```
